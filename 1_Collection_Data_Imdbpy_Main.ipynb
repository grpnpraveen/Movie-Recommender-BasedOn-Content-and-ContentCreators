# Step 1 
Collecting only single valued data for each movie

```python
import imdb
from tqdm import tqdm
import pandas as pd
# wanted_list=["year","title","rating","runtimes","plot","cover url","full-size cover url"]
movies_db=imdb.IMDb()
movies_csv=pd.read_csv("IMDbmovies.csv",dtype='str')
ids=movies_csv["imdb_title_id"]
string_class=""
# item lists 
ids_list=[]
year_list=[]
title_list=[]
genre_list=[]
languages_list=[]
rating_list=[]
votes_list=[]
runtime_list=[]
countries_list=[]
directorID_list=[]
writerID_list=[]
castID_list=[]
plot_list=[]
smallcover_list=[]
fullsizecover_list=[]
nothing="None"
count=1
#start
for i in tqdm(ids):
    id=i[2:]
    movie=movies_db.get_movie(id)
    ids_list.append(i)
    try:
        year_list.append(movie["year"])
    except:
        year_list.append(nothing)
    try:
        title_list.append(movie["title"])
    except:
        title_list.append(nothing)
    #geners
    try:
        genres=movie["genres"]
        x=""
        l=len(genres)
        for j in range(l):
            x=x+genres[j]
            if(j!=(l-1)):
                x=x+"|"
        genre_list.append(x)
    except:
        genre_list.append(nothing)
    try:
        rating_list.append(movie["rating"])
    except:
        rating_list.append(nothing)
    try:
        votes_list.append(movie["votes"])
    except:
        votes_list.append(nothing)
    try:
        runtime_list.append(movie["runtimes"][0])
    except:
        runtime_list.append(nothing)
    try:
        #countries
        countries=movie["countries"]
        country_a=""
        l_countries=len(countries)
        for country in range(l_countries):
            country_a=country_a+countries[country]
            if(country!=(l_countries-1)):
                country_a=country_a+"|"
        countries_list.append(country_a)
    except:
        countries_list.append(nothing)
    try:
        #languages
        languages=movie["languages"]
        languages_l=len(languages)
        language_s=""
        for language in range(languages_l):
            language_s=language_s+languages[language]
            if(language!=(languages_l-1)):
                language_s=language_s+"|"
        languages_list.append(language_s)
    except:
        languages_list.append(nothing)
    try:
        #directed by
        directors=movie["directors"]
        director_id=""
        l_directors=len(directors)
        for director in range(l_directors):
            CHECK=directors[director].getID()
            if(type(CHECK)==type(string_class)):
                director_id=director_id + CHECK
                if(director!=(l_directors-1)):
                    director_id=director_id+"|"
        directorID_list.append(director_id)
    except:
        directorID_list.append(nothing)
    try:
        #writer
        writers=movie["writers"]
        writer_id=""
        l_writers=len(writers)
        for writer in range(l_writers):
            CHECK=writers[writer].getID()
            if(type(CHECK)==type(string_class)):
                writer_id=writer_id+CHECK
                if(writer!=(l_writers-1)):
                    writer_id=writer_id+"|"
        writerID_list.append(writer_id)
    except:
         writerID_list.append(nothing)
    try:
        #cast
        casts=movie["cast"]
        cast_id=""
        l_cast=len(casts)
        for each in range(l_cast):
            CHECK=casts[each].getID()
            if(type(CHECK)==type(string_class)):
                cast_id=cast_id+CHECK
                if(each!=(l_cast-1)):
                    cast_id=cast_id+"|"
        castID_list.append(cast_id)
    except:
        castID_list.append(nothing)
    try:
        #plot
        plot=movie["plot"][0]
        mark_start=plot.find("::")
        plot_list.append(plot[:mark_start])
    except:
           plot_list.append(nothing)
    try:
        smallcover_list.append(movie["cover url"])
    except:
        smallcover_list.append(nothing)
    try:
        fullsizecover_list.append(movie["full-size cover url"])
    except:
         fullsizecover_list.append(nothing)
    count=count+1
    if(count==5):
        break
#end
final_info=pd.DataFrame(
            {
                "id":ids_list,
                "year":year_list,
                "Movie":title_list,
                "genre":genre_list,
                "languages":languages_list,
                "rating":rating_list,
                "votings":votes_list,
                "duration":runtime_list,
                "Countries":countries_list,
                "directedBy":directorID_list,
                "writers":writerID_list,
                "cast":castID_list,
                "plot":plot_list,
                "small cover":smallcover_list,
                "fullsize cover":fullsizecover_list
            }
            )
print(final_info)

final_info.to_csv("result1.csv")
```
